```python
def get_sum(n1 = 0, n2 = 0): # params
    return n1 + n2

# print(get_sum(10, 20)) # arguments
# print(get_sum(10))
# print(get_sum())
# print(get_sum(10, 20, 30))
```

```python
# *arguments -> *args -> don't know the no of inputs
def get_sum(*nums): # params
    # print(nums) # tuple
    total = 0
    for num in nums:
        total += num
    return total

print(get_sum(10, 20)) # arguments
print(get_sum(10))
print(get_sum())
print(get_sum(10, 20, 30))
```

```python
def welcome(name = 'Guest', roll_no = -1):
    return f'Welcome to Intellipaat {name}. Your ID is {roll_no}'

# print(welcome('vasi', 10))
# print(welcome('vasi'))
# print(welcome())
# print(welcome(roll_no = 10, name = 'vasi', place = 'chennai'))
```

```python
# **keyword arguments -> **kwargs
def welcome(**data):
    # print(data) # dict
    return f'Welcome to Intellipaat {data.get('name', 'Guest')}. Your ID is {data.get('roll_no', -1)}'

print(welcome(roll_no = 10, name = 'vasi', place = 'chennai'))
print(welcome(roll_no = 10))
print(welcome())
```
